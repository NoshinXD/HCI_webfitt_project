## Making our overall files

```python
parent_path = '/content/drive/MyDrive/my_grad_HCI/HCI_project/data/extracted_data/'

for i in range(4):
    p = i + 1
    # print('p', p)

    # if p > 1:
    #     break

    for j in range(4):
        c = j + 1
        # print('c', c)

        # if c > 1:
        #     break

        for k in range(3):
            s = k + 1
            # print('s', s)

            # if s > 1:
            #     break

            substring = '/WebFitts_P' + str(p) + '_S' + str(s) + '_C' + str(c) + '_Touchpad' # /WebFitts_P1_S1_C1_Touchpad
            click_filename = parent_path + 'P' + str(p) + substring + substring + '_click.csv'
            overall_filename = parent_path + 'P' + str(p) + substring + substring + '_overall.csv'

            click_df = pd.read_csv(click_filename)
            overall_df = pd.read_csv(overall_filename)

            # print(click_df.columns)
            # print(overall_df.columns)

            overall_df.loc[0, 'Number of Targets'] = click_df.loc[0, 'Number of Targets']

            my_overall_filename = parent_path + 'P' + str(p) + substring + substring + '_our_overall.csv'
            overall_df.to_csv(my_overall_filename, index=False)

print('done!')
```

```python
common_path = '/content/drive/MyDrive/my_grad_HCI/HCI_project/data/extracted_data/'

temp_filename = common_path + 'P1/WebFitts_P1_S1_C1_Touchpad/WebFitts_P1_S1_C1_Touchpad_our_overall.csv'
temp_df = pd.read_csv(temp_filename)
merged_df = pd.DataFrame(columns=temp_df.columns)

for i in range(4):
    p = i + 1

    # if p == 2:
    #     break

    for j in range(4):
        c = j + 1

        # if c == 2:
        #     break

        for k in range(3):
            s = k + 1

            # if s == 2:
            #     break

            # filename = '/u/epw9kz/academic/first_year/HCI/extracted_data/P1/WebFitts_P1_S1_C1_Touchpad/WebFitts_P1_S1_C1_Touchpad_click.csv'
            # filename = '/content/drive/MyDrive/my_grad_HCI/HCI_project/data/extracted_data/P1/WebFitts_P1_S1_C1_Touchpad/WebFitts_P1_S1_C1_Touchpad_overall.csv'
            substring = '/WebFitts_P' + str(p) + '_S' + str(s) + '_C' + str(c) + '_Touchpad' # /WebFitts_P1_S1_C1_Touchpad
            filename = common_path + 'P' + str(p) + substring + substring + '_our_overall.csv'
            # print(filename)

            temp_df = pd.read_csv(filename)
            frames = [merged_df, temp_df]

            merged_df = pd.concat(frames)

print(merged_df.shape)

merged_df = merged_df[['Participant Code', 'Session Code', 'Condition Code', 'Hand Dominance',
                       'Pointing Device', 'Device Experience', 'Number of Targets',
                       'Mean Completion Time (ms)', 'Mean Click Error (%)', 'Mean Throughput (bps)']]


output_filename = common_path + 'all_overall_data_arranged.csv'
merged_df.to_csv(output_filename, index=False)
print('done!')
```
